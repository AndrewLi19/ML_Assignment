```python
import pandas as pd
import numpy as np
x=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/train_features.csv')
x.to_numpy()
y=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/train_target.csv')
y.to_numpy()
x_hat=np.concatenate((x,np.ones((x.shape[0],1))),axis=1)
y=y.values
```

```python
def delta(beta):
    a=np.zeros(11)
    b=np.zeros((11,11))
    for i in range(x_hat.shape[0]):
        p1=np.exp(np.dot(x_hat[i],beta))/(1+np.exp(np.dot(x_hat[i],beta)))
        a-=(y[i]-p1)*x_hat[i]
        b+=p1*(1-p1)*np.dot(np.mat(x_hat[i]).T,np.mat(x_hat[i]))
    d=np.dot(np.linalg.pinv(b),a)
    return d

beta=np.ones(11)
d=delta(beta)
while np.any(abs(d)>0.0001):
    beta=beta-d
    d=delta(beta)
```

```python
val_features=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/val_features.csv')
val_features.to_numpy()
val_features=np.concatenate((val_features,np.ones((val_features.shape[0],1))),axis=1)
val_target=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/val_target.csv')
val_target=val_target.values
theta=np.arange(0.0,1.0,0.1)
for m in range(0,10):
    tp=0
    fp=0
    fn=0
    tn=0
    for i in range(1,val_features.shape[0]):
        z=np.dot(np.matrix(val_features[i]),beta)
        f=1/(1+np.exp(-z))
        if f>=theta[m]:
            tag=1
        if f<theta[m]:
            tag=0
        if tag==1 and val_target[i]==1:
            tp+=1
        if tag==1 and val_target[i]==0:
            fp+=1
        if tag==0 and val_target[i]==1:
            fn+=1
        if tag==0 and val_target[i]==0:
            tn+=1
    accuracy=(tp+tn)/(tp+fp+fn+tn)
    precision=tp/(tp+fp)
    recall=tp/(tp+fn)
    print(theta[m],accuracy,precision,recall)
```

```python
test_feature=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/test_feature.csv')
test_feature=np.concatenate((test_feature,np.ones((test_feature.shape[0],1))),axis=1)
tag=list(range(test_feature.shape[0]))
for i in range(0,test_feature.shape[0]):
    z=np.dot(np.matrix(test_feature[i]),beta)
    f=1/(1+np.exp(-z))
    if f>=0.5:
        tag[i]=1
    if f<0.5:
        tag[i]=0
df_out = pd.DataFrame({'class':tag})
df_out.to_csv("./58120121_1.csv",index=False,sep=',')
```
