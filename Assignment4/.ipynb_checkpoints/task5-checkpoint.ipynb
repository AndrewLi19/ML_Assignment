```python
from scipy.stats import ttest_rel, f
import numpy as np
```

```python
a=[[2,5,4,2,3],[3,4,5,3,4],[1,2,1,1,1],[5,3,2,5,5],[4,1,3,4,2]]
b=[3.2,3.8,1.2,4,2.8]
```

```python
def f_test(x,y):
    var1 = np.var(x, ddof=1)
    var2 = np.var(y, ddof=1)
    F = max(var1 / var2,var2 / var1)
    if F>3.007:
        return False
    else:
        return True

def n_test(x,y):
    tmp=abs(x-y)
    if tmp>2.728:
        return False
    else:
        return True
```

```python
for m in range(5):
    for n in range(5):
        tmp_1=f_test(a[m],a[n])
        if tmp_1:
            print("算法",m+1,"和算法",n+1,"性能相同")
        else:
            print("算法",m+1,"和算法",n+1,"性能不同")
            tmp_2=n_test(b[m],b[n])
            if tmp_2:
                print("------算法",m+1,"和算法",n+1,"无显著差异")
            else:
                print("------算法",m+1,"和算法",n+1,"有显著差异")
```
