```python
import pandas as pd
import numpy as np
```

```python
x=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/train_features.csv')
x.to_numpy()
y=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/train_target.csv')
y.to_numpy()
```

```python
x_hat=np.concatenate((x,np.ones((x.shape[0],1))),axis=1)
beta=np.dot(np.dot(np.matrix(np.dot(x_hat.T,x_hat)).I,x_hat.T),y)
beta
```

```python
val_features=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/val_features.csv')
val_features.to_numpy()
val_features=np.concatenate((val_features,np.ones((val_features.shape[0],1))),axis=1)
val_target=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/val_target.csv')
val_target=val_target.values
tp=0
fp=0
fn=0
tn=0
for i in range(1,val_features.shape[0]):
    z=np.dot(np.matrix(val_features[i]),beta)
    f=1/(1+np.exp(-z))
    if f>=0.5:
        tag=1
    if f<0.5:
        tag=0
    if tag==1 and val_target[i]==1:
        tp+=1
    if tag==1 and val_target[i]==0:
        fp+=1
    if tag==0 and val_target[i]==1:
        fn+=1
    if tag==0 and val_target[i]==0:
        tn+=1
accuracy=(tp+tn)/(tp+fp+fn+tn)
precision=tp/(tp+fp)
recall=tp/(tp+fn)
print(accuracy,precision,recall)
```
