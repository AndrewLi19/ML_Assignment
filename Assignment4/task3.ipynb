```python
import pandas as pd
import numpy as np
x=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/train_features.csv')
x.to_numpy()
y=pd.read_csv('./data/ML4_programming-all/ML4_programming-all/train_target.csv')
y.to_numpy()
x_hat=np.concatenate((x,np.ones((x.shape[0],1))),axis=1)
y=y.values
```

```python
def delta(beta):
    a=np.zeros(11)
    b=np.zeros((11,11))
    for i in range(x_hat.shape[0]):
        p1=np.exp(np.dot(x_hat[i],beta))/(1+np.exp(np.dot(x_hat[i],beta)))
        a-=(y[i]-p1)*x_hat[i]
        b+=p1*(1-p1)*np.dot(np.mat(x_hat[i]).T,np.mat(x_hat[i]))
    d=np.dot(np.linalg.pinv(b),a)
    return d

b=np.ones(11)
d=delta(b)
while np.any(abs(d)>0.0001):
    b=b-d
    d=delta(b)
```
